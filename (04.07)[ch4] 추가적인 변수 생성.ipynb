- Age    
> 20~60 working year / 나머지    
연령대 + 성별

```python
# 20~60 경제활동 / 나머지
def active_age(age) :
  active_age = ''
  if age >= 20 and age <= 60  : active_age = 'active'
  else : active_age = 'inactive'
  return active_age
```

```python
df['Active_age'] = df['Age'].apply(lambda x : active_age(x))
```

```python
df['Active_age']
```

```python
def set_m_age(age) :
  set_gender_age = ''
  if age <10 :
    set_gender_age = '0M'
  elif age <20 : set_gender_age = '10M'
  elif age <30 : set_gender_age = '20M'
  elif age <40 : set_gender_age = '30M'
  elif age <50 : set_gender_age = '40M'
  elif age <60 : set_gender_age = '50M'
  elif age <70 : set_gender_age = '60M'
  elif age <80 : set_gender_age = '70M'
  elif age <90 : set_gender_age = '80M'
  else : set_gender_age = '90M'

  return set_gender_age

def set_f_age(age) :
  set_gender_age = ''
  if age < 10 : set_gender_age = '0F'
  elif age < 20 : set_gender_age = '10F'
  elif age < 30 : set_gender_age = '20F'
  elif age < 40 : set_gender_age = '30F'
  elif age < 50 : set_gender_age = '40F'
  elif age < 60 : set_gender_age = '50F'
  elif age < 70 : set_gender_age = '60F'
  elif age < 80 : set_gender_age = '70F'
  elif age < 90 : set_gender_age = '80F'
  else : set_gender_age = '90F'

  return set_gender_age
```

```python
df['Set_gender_age'] = df['Age'].copy()
```

```python
con1 = df['Gender'] == 'M'
df.loc[con1,'Set_gender_age'] = df.loc[con1,'Set_gender_age'].apply(lambda x : set_m_age(x))
```

```python
con2 = df['Gender'] == 'F'
df.loc[con2,'Set_gender_age'] = df.loc[con2,'Set_gender_age'].apply(lambda x : set_f_age(x))
```

```python
# 1) 파트 / 풀 나머지m
emp_list = ['Children or Armed Forces', 'Full-Time', 'Not Working', 'Choice Part-Time',
           'Part-Time (Usually Part-Time)', 'Part-Time (Usually Full-Time)','Seeking Full-Time', 'Seeking Part-Time']


for i in range(len(emp_list)):
  if i == 0 :
    df.loc[(df['Employment_Status'] == emp_list[i]), 'Group1_Employment'] = 'other'
  elif i == 1 :
    df.loc[(df['Employment_Status'] == emp_list[i]), 'Group1_Employment'] = 'full_time'
  elif i == 2 :
    df.loc[(df['Employment_Status'] == emp_list[i]), 'Group1_Employment'] = 'other'
  elif 2 < i <= 5 :
    df.loc[(df['Employment_Status'] == emp_list[i]), 'Group1_Employment'] = 'part_time'
  elif i > 5:
    df.loc[(df['Employment_Status'] == emp_list[i]), 'Group1_Employment'] = 'other'
```

- Citizenship
> Native / Foreign

```python
# Citizenship -> Native인지 아닌지 구분
def native_ox(df) :
  con1 = df['Citizenship'] == 'Native'
  con2 = df['Citizenship'] == 'Native (Born in Puerto Rico or US Outlying)'
  con3 = df['Citizenship'] == 'Native (Born Abroad)'
  con4 = df['Citizenship'] == 'Foreign-born (Non-US Citizen)'
  con5 = df['Citizenship'] == 'Foreign-born (Naturalized US Citizen)'
  df.loc[con1 | con2 | con3,'Native_ox'] = 'Native'
  df.loc[con4 | con5,'Native_ox'] = 'Non-Native'
```

```python
native_ox(df)
```

```python
df['Native_ox'].unique()
```

- Hispanic Origin
> All other O / X

```python
# Hispanic -> All other인지 아닌지 구분
def allother_ox(df) :
  con0 = df['Hispanic_Origin'] == 'All other'
  con1 = df['Hispanic_Origin'] != 'All other'
  df.loc[con0,'Allother_ox'] = 'O'
  df.loc[con1,'Allother_ox'] = 'X'
```

```python
allother_ox(df)
```
